```python
from kafka import KafkaProducer
import json
import random
import time

# Configuracin del broker Kafka (ajusta IP si es necesario)
KAFKA_BROKER = 'localhost:29092'
TOPIC = 'sensores'

# Crear productor
producer = KafkaProducer(
    bootstrap_servers=KAFKA_BROKER,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Funcin para generar datos simulados
def generar_dato_sensor():
    return {
        'sensor_id': random.randint(100, 999),
        'temperatura': round(random.uniform(20.0, 100.0), 2),
        'humedad': round(random.uniform(30.0, 90.0), 2),
        'timestamp': int(time.time())
    }

# Enviar datos cada 2 segundos
try:
    while True:
        mensaje = generar_dato_sensor()
        print(f"Enviando: {mensaje}")
        producer.send(TOPIC, mensaje)
        time.sleep(2)
except KeyboardInterrupt:
    print("Interrumpido. Cerrando productor.")
    producer.close()
```

```python
from kafka import KafkaProducer, KafkaConsumer
import json
import time

# Configuracin
KAFKA_BROKER = 'localhost:29092'
TOPIC = 'sensores'

# -----------------------
# PRODUCTOR DE PRUEBA
# -----------------------
producer = KafkaProducer(
    bootstrap_servers=KAFKA_BROKER,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Enviar un solo mensaje de prueba
mensaje = {
    'sensor_id': 123,
    'temperatura': 55.5,
    'humedad': 70.2,
    'timestamp': int(time.time())
}

producer.send(TOPIC, mensaje)
producer.flush()
print(f"Mensaje enviado al topic '{TOPIC}': {mensaje}")
producer.close()

# -----------------------
# CONSUMIDOR DE PRUEBA
# -----------------------
consumer = KafkaConsumer(
    TOPIC,
    bootstrap_servers=KAFKA_BROKER,
    auto_offset_reset='earliest',
    group_id='test-consumer-group',
    enable_auto_commit=True,
    value_deserializer=lambda m: json.loads(m.decode('utf-8'))
)

print(f"\nEsperando mensajes en el topic '{TOPIC}'... (Ctrl+C para salir)\n")
for msg in consumer:
    print(f"Mensaje recibido: {msg.value}")
    break  # Salir despus del primer mensaje recibido

consumer.close()
```
